```python
# import the train data

from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['target','flower_name'],axis = 'columns')  # drop a target and flower_name columns and create a feature
```

```python
x
```

```python
y = df.target
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1                                                 )
```

```python
# import the k nearest classifire

from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)  # check accuracy
```

```python
knn.predict([[4.8,5,1.6,0.2]])
```

```python
# to create a confusion metrix

from sklearn.metrics import confusion_matrix
```
